```python
import cv2
import requests
import matplotlib.pyplot as plt
from deepface import DeepFace
```

```python
img = cv2.imread("/content/images.jpeg")
plt.imshow(img[:,:,::-1])
plt.show()
```

```python
result = DeepFace.analyze(img, actions = ['emotion'])
print(result)
```

```python
query = str(max(zip(result[0]['emotion'].values(),result[0]['emotion'].keys()))[1])
print(query)
```

```python
if query == 'happy1':
  song_name1='oru kal oru kannadi'
  print(song_name1)
elif query == 'sad1':
  song_name1='until i found her'
  print(song_name1)
elif query == 'angry1':
  song_name1='Ding Dong'
  print(song_name1)
```

```python
# Install spotipy
!pip install spotipy

import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
from IPython.display import display, IFrame

# Set your Spotify credentials
client_id = 'a7df7cd7bf164810bb3eebc1d968d76f'
client_secret = '6a08958fe28a47698691f7746c96a4b7'

# Authenticate using the Client Credentials flow
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=client_id, client_secret=client_secret))

# Function to search for a track and return its information
def get_track_info(song_name):
    results = sp.search(q=song_name, limit=1, type='track')
    if results['tracks']['items']:
        track = results['tracks']['items'][0]
        return {
            'name': track['name'],
            'artist': track['artists'][0]['name'],
            'album': track['album']['name'],
            'release_date': track['album']['release_date'],
            'external_url': track['external_urls']['spotify']
        }
    else:
        return None

# Map emotions to song names
def get_song_name_for_emotion(emotion):
    if emotion == 'happy1':
        return 'oru kal oru kannadi'
    elif emotion == 'sad1':
        return 'until i found her'
    elif emotion == 'angry1':
        return 'Ding Dong'
    else:
        return None

# Query example
query = 'sad1'  # Change this to 'happy1', 'sad1', or 'angry1' to test different queries

# Get the song name based on the query
song_name = get_song_name_for_emotion(query)

# Get the track info from Spotify
if song_name:
    track_info = get_track_info(song_name)
    if track_info:
        print(f"Track Name: {track_info['name']}")
        print(f"Artist: {track_info['artist']}")
        print(f"Album: {track_info['album']}")
        print(f"Release Date: {track_info['release_date']}")
        print(f"Listen on Spotify: {track_info['external_url']}")
        # Display the track in an iframe
        display(IFrame(src=track_info['external_url'], width="300", height="80"))
    else:
        print("Track not found on Spotify.")
else:
    print("Invalid emotion query.")
```
